# Essential Linear Algebra Concepts

```python
import numpy as np

# Vectors
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

# Vector addition and subtraction
v_add = v1 + v2
v_sub = v1 - v2
print("Vector addition:", v_add)
print("Vector subtraction:", v_sub)

# Scalar multiplication
v_scalar = 3 * v1
print("Scalar multiplication:", v_scalar)

# Matrices
m1 = np.array([[1, 2], [3, 4]])
m2 = np.array([[5, 6], [7, 8]])

# Matrix addition and subtraction
m_add = m1 + m2
m_sub = m1 - m2
print("Matrix addition:\n", m_add)
print("Matrix subtraction:\n", m_sub)

# Matrix multiplication
m_mult = np.dot(m1, m2)
print("Matrix multiplication:\n", m_mult)

# Dot product
dot_prod = np.dot(v1, v2)
print("Dot product:", dot_prod)
```

# Gradient Descent Optimization

```python
import matplotlib.pyplot as plt

# Define the quadratic function and its gradient
def f(x):
    return x**2

def gradient(x):
    return 2*x

# Gradient descent parameters
x = 10  # Initial guess
learning_rate = 0.1
iterations = 20

# Lists to store the values for plotting
x_values = [x]
f_values = [f(x)]

# Gradient descent loop
for _ in range(iterations):
    grad = gradient(x)
    x = x - learning_rate * grad
    x_values.append(x)
    f_values.append(f(x))

# Plot the function and the optimization path
x_range = np.linspace(-10, 10, 400)
plt.plot(x_range, f(x_range), label='$f(x) = x^2$')
plt.scatter(x_values, f_values, color='red')
plt.plot(x_values, f_values, color='red', label='Gradient Descent Path')
plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.title('Gradient Descent Optimization')
plt.legend()
plt.grid(True)
plt.show()
```

# Hands-on

```python
# Define the quadratic function and its gradient
def f(x):
    return (x-3)**2 + 4

def gradient(x):
    return 2*(x-3)

# Gradient descent parameters
initial_values = [10, -10, 5, -5]  # Different initial guesses
learning_rate = 0.1
iterations = 20

# Plot the function
x_range = np.linspace(-10, 10, 400)
plt.plot(x_range, f(x_range), label='$f(x) = (x-3)^2 + 4$')

for x in initial_values:
    x_values = [x]
    f_values = [f(x)]
    for _ in range(iterations):
        grad = gradient(x)
        x = x - learning_rate * grad
        x_values.append(x)
        f_values.append(f(x))
    plt.scatter(x_values, f_values)
    plt.plot(x_values, f_values, label=f'Path from initial x={x_values[0]}')

plt.xlabel('x')
plt.ylabel('$f(x)$')
plt.title('Gradient Descent Optimization from Different Initial Values')
plt.legend()
plt.grid(True)
plt.show()
```
